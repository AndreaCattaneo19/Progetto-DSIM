Notebook utilizzato per la registrazione delle tracce audio, successivamente utilizzate per la classificazione.

```python
import sounddevice as sd
from scipy.io.wavfile import write
import os
```

Sono state registrate e salvate per ciascuna classe 50 tracce audio da 5 secondi l'una.

```python
duration = 5 # (seconds)
rec_rate = 44100
```

```python
def registratore(nome, folder):
        for x in range (0, 50):
            rec = sd.rec(int(duration * rec_rate), samplerate=rec_rate, channels=1, blocking=True)
            write(str(folder) + str(nome) + str(x) + '.wav', rec_rate, rec)
```

```python
rimbalzino_folder= './rimbalzino_folder/'
bigmari_folder= './bigmari_folder/'
stemanzo_folder='./stemanzo_folder/'
```

##Registrazione Audio

### Rimbalzino

```python
registratore('rimbalzino', rimbalzino_folder)
```

### Bigmari

```python
registratore('bigmari', bigmari_folder)
```

### SteManzo

```python
registratore('stemanzo', stemanzo_folder)
```
